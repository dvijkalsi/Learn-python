```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
style.use('dark_background')
```

```python
x = [2,4,6]
y = [4,8,24]
x2 = [3,6,9]
y2 = [10,15,38]
plt.plot(x,y)
plt.plot(x2,y2)

plt.show()
```

```python
style.use('dark_background')
plt.title('Epic Chart')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```
