```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib import style
```

```python
database = sb.load_dataset('diamonds')
print(database.head())
```

```python
sb.distplot(database['carat'],bins=1)
plt.show()
```

```python
database = sb.load_dataset('tips')
print(database.head())
```

```python
sb.jointplot(x='total_bill',y='tip',data=database)
plt.show()
```

```python
database = sb.load_dataset('flights')
print(database)
```

```python
sb.catplot(x='month',y='passengers',data=database,kind='box')
plt.show()
```
